```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris =   pd.read_csv("/content/Iris.csv")
print(iris.head())
```

```python
print(iris.describe())
```

```python
iris.plot(kind="scatter", x="SepalLengthCm",   y="SepalWidthCm")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris_data=load_iris()
```

```python
iris_df = pd.DataFrame(iris_data.data, columns = iris_data.feature_names) #creating dataframe
```

```python
kmeans = KMeans(n_clusters=3) #Applying Kmeans classifier
```

```python
x = iris.iloc[:, :-1].values #last column values excluded
```

```python
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
print(kmeans.cluster_centers_) #display cluster centers
```

```python
plt.scatter(x[y_kmeans   == 0, 0], x[y_kmeans == 0, 1],s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans   == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans   == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')   #Visualising the clusters - On the first two columns
plt.scatter(kmeans.cluster_centers_[:,   0], kmeans.cluster_centers_[:,1],s = 100, c = 'black', label = 'Centroids',marker='s')   #plotting the centroids of the clusters

plt.legend()
plt.show()
```

```python
plt.scatter(x[y_kmeans   == 0, 1], x[y_kmeans == 0, 2],s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans   == 1, 1], x[y_kmeans == 1, 2],s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans   == 2, 1], x[y_kmeans == 2, 2],s = 100, c = 'green', label = 'Iris-virginica')   #Visualising the clusters - On 2nd & 3rd columns
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,2],s = 100, c = 'black', label = 'Centroids',marker='s')   #plotting the centroids of the clusters

plt.legend()
plt.show()
```
